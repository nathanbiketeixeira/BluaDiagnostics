```python
SYSTEM_PROMPT = """
Você é o BluaDiagnostics, assistente virtual da Care Plus.

Você pode:
- orientar triagem;
- sugerir teleconsulta;
- identificar sinais de alerta.

Você NÃO pode:
- diagnosticar;
- prescrever;
- substituir médicos.
"""
```

```python
from openai import OpenAI
from google.colab import userdata

client = OpenAI(
    api_key=userdata.get('OPENAI_API_KEY')
)
```

```python
messages = [
    {
        "role": "system",
        "content": SYSTEM_PROMPT
    }
]
```

```python
TOOLS = [
    {
        "type": "function",
        "function": {
            "name": "agendar_teleconsulta",
            "description": "Agenda teleconsulta",
            "parameters": {
                "type": "object",
                "properties": {
                    "especialidade": {
                        "type": "string"
                    }
                },
                "required": ["especialidade"]
            }
        }
    }
]
```

```python
def agendar_teleconsulta(especialidade):
    return {
        "status": "sucesso",
        "especialidade": especialidade,
        "horario": "amanhã às 14h"
    }
```

```python
messages.append({
    "role": "user",
    "content": "Quero falar com um psicólogo."
})
```

```python
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages,
    tools=TOOLS,
    tool_choice="auto"
)

message = response.choices[0].message

print(message)
```

```python
import json

if message.tool_calls:

    tool_call = message.tool_calls[0]

    args = json.loads(tool_call.function.arguments)

    resultado = agendar_teleconsulta(
        especialidade=args["especialidade"]
    )

    print(resultado)
```
